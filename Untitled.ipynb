```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 
```

```python
data = pd.read_csv('EgyptianLeague.csv')
data.head()
```

```python
import itertools
data['Total Goals'] = data['Home Goals']+data['Away Goals']
sum_goals=data.groupby('Season').sum()['Total Goals']
mean_goals=data.groupby('Season').mean()['Total Goals']
std_goals=data.groupby('Season').std()['Total Goals']
max_goals=data.groupby('Season').max()['Total Goals']
min_goals=data.groupby('Season').min()['Total Goals']

tot_gls = pd.concat([sum_goals, mean_goals, std_goals, max_goals, min_goals], axis=1)
tot_gls.columns=['Goals','Mean','Std','Max','Min']
cols = [ 'Goals', 'Mean', 'Std', 'Max','Min' ]
length = len(cols)
cs   = ["r","g","b","c",'y'] 
plt.figure(figsize=(12,16))

for i,j,k in itertools.zip_longest(cols,range(length),cs):
    plt.subplot(length,length/length,j+1)
    sns.pointplot(tot_gls.index,tot_gls[i],color=k)
    plt.subplots_adjust(hspace =.3)
```

#### Season 2011 has highest mean and second minimum goals !?
Season 2011 second minimum in the number of matches as well with only 143, maybe it could be (if the season continued) one of the highest season in goals, or maybe there's outliers that affected the mean like this (we will investigate that further in box plot)

### The Top ten scoring teams in all seasons  

```python

x=data.groupby('Home Team')['Home Goals'].sum().reset_index().sort_values(by='Home Goals',ascending=False).rename(columns={'Home Team':'Team','Home Goals':'Goals'})
y= data.groupby('Away Team')['Away Goals'].sum().reset_index().sort_values(by='Away Goals',ascending=False).rename(columns={'Away Team':'Team','Away Goals':'Goals'})
xy = pd.concat([x,y],axis=0)
all_goals = xy.groupby('Team')['Goals'].sum().reset_index().sort_values(by='Goals',ascending=False)[:10]
fig = plt.figure(figsize=(8,10))
ax=sns.barplot(x='Goals',y='Team',data=all_goals)
for i,j in enumerate(all_goals['Goals']):
    ax.text(.5,i,j,weight='bold')
plt.show()
```

**The most scoring teams was el Ahly, Zamalek and Masr el makasa, but because not all teams participated in all 8 seasons, we will try to investigate more to see if there's a team or two that were shining in short time whether they came from devision 1 or regualted**

```python
x=data.groupby(['Home Team','Season'])['Home Goals'].sum().reset_index().rename(columns={'Home Team':'Team','Home Goals':'Goals'})
y=data.groupby(['Away Team','Season'])['Away Goals'].sum().reset_index().rename(columns={'Away Team':'Team','Away Goals':'Goals'})
# xp = x.groupby('Season')
# y= data.groupby('Away Team')['Away Goals'].sum().reset_index().sort_values(by='Away Goals',ascending=False).rename(columns={'Away Team':'Team','Away Goals':'Goals'})
xy = pd.concat([x,y],axis=0)
all_goals = xy.groupby(['Team','Season'])['Goals'].sum().reset_index().sort_values(by='Goals',ascending=False)
plt.figure(figsize=(12,15))
sns.barplot(y='Team',x='Goals',hue='Season',data=all_goals[:20])
plt.show()
```

```python
plt.figure(figsize=(12,8))
home = data.groupby('Home Team')['Home Goals'].sum().reset_index().sort_values(by='Home Goals',ascending=False)[:10]
away = data.groupby('Away Team')['Away Goals'].sum().reset_index().sort_values(by='Away Goals',ascending=False)[:10]
plt.subplot(121)
ax=sns.barplot(x='Home Goals',y='Home Team',data=home)
for i,j in enumerate(home['Home Goals']):
    ax.text(0.5,i,j,weight='bold')
plt.ylabel("")
plt.subplot(122)
plt.subplots_adjust(wspace=.5)
ax=sns.barplot(x='Away Goals',y='Away Team',data=away)
for i,j in enumerate(away['Away Goals']):
    ax.text(0.5,i,j,weight='bold')

```

```python
tt=data.groupby(['Season','Home Team']).mean()['Home Goals'].reset_index().sort_values(by='Home Goals',ascending=False).rename(columns={'Home Team':'Team','Home Goals':'Goals'})
tt.groupby('Team')['Goals'].mean().reset_index().sort_values(by='Goals',ascending=False)
yy=data.groupby(['Season','Away Team']).mean()['Away Goals'].reset_index().sort_values(by='Away Goals',ascending=False).rename(columns={'Away Team':'Team','Away Goals':'Goals'})
yy.groupby('Team')['Goals'].mean().reset_index().sort_values(by='Goals',ascending=False)
tot = pd.concat([tt,yy],axis=0)
toty=tot.groupby(['Team']).mean()['Goals'].reset_index().sort_values(by='Goals',ascending=False)[:20]
plt.figure(figsize=(12,8))
ax=sns.barplot(x='Goals',y='Team',data=toty)
for i,j in enumerate(toty['Goals']):
    ax.text(.1,i,j,weight='bold')
plt.show()
```

```python
hperm=data.groupby('Home Team').agg({'Home Goals':'sum','Fixture':'count'}).reset_index().rename(columns={'Home Team':'Team',"Home Goals":"Goals"})
aperm = data.groupby('Away Team').agg({'Away Goals':'sum','Fixture':'count'}).reset_index().rename(columns={'Away Team':'Team',"Away Goals":'Goals'})
perm=pd.concat([hperm,aperm],axis=0).groupby('Team')[['Goals','Fixture']].sum().reset_index()
perm=perm.sort_values(by='Goals',ascending=True)
plt.figure(figsize=(12,8))
plt.scatter(x=perm['Fixture'],y=perm['Team'],s=perm['Goals'],alpha=.7)
plt.xlabel('Matches Played')
plt.show()
```

```python
pp=pd.concat([hperm,aperm],axis=0).groupby('Team')[['Goals','Fixture']].sum().sort_values(by='Goals',ascending=True)
pp.plot(kind='barh',stacked =True,figsize=(12,10))
plt.show()
```

```python
pp['Fixture'].plot.pie(autopct="%0.1f%%",figsize=(15,15))
my_circ = plt.Circle((0,0),.7,color ="white")
plt.gca().add_artist(my_circ)
plt.title('Proportion of Matches played')
plt.show()
```
